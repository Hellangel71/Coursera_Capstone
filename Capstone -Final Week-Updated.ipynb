# Define a function to obtain co-ordinates of Pune Neighbourhoods

```python
# define a function to get coordinates
def get_latlng(neighbourhood):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Pune, India'.format(neighbourhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
coords = [ get_latlng(neighbourhood) for neighbourhood in pune_df["Neighbourhoods"].tolist() ]
```

# Search for venues

```python
radius = 2000
LIMIT = 100

venues = []

for lat, long, neighbourhood in zip(pune_df['Latitude'], pune_df['Longitude'], pune_df['Neighbourhoods']):
    
    # create the API request UuRL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            neighbourhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

# Create a dataframe of these obtained venues

```python
# creating a new dataframe
df_venues = pd.DataFrame(venues)

# Assign new column names for the dataframe
df_venues.columns = ['Neighbourhoods', 'Latitude', 'Longitude', 'Venue', 'V_Latitude', 'V_Longitude', 'Category']

df_venues.head()
```

# One-Hot-Encoding

```python
# one hot encoding
pune_onehot = pd.get_dummies(df_venues[['Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
pune_onehot['Neighbourhoods'] = df_venues['Neighbourhoods'] 

# move neighborhood column to the first column
fixed_columns = [pune_onehot.columns[-1]] + list(pune_onehot.columns[:-1])
pune_onehot = pune_onehot[fixed_columns]

print(pune_onehot.shape)
pune_onehot.head()
```

# Obtain some statistical venue information

```python
pune_grouped = pune_onehot.groupby(["Neighbourhoods"]).mean().reset_index()

print(pune_grouped.shape)
pune_grouped
```

# Filter only Multiplex venues

```python
len(pune_grouped[pune_grouped["Multiplex"] > 0])
```

```python
pune_mul = pune_grouped[["Neighbourhoods","Multiplex"]]
pune_mul
```

# Use K-Means Clustering

```python
# set number of clusters
kclusters = 5

kl_cluster = pune_mul.drop(["Neighbourhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(kl_cluster)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]

```

# Create a dataframe with cluster labels

```python
pune_merged = pune_mul.copy()
pune_merged["Cluster Labels"] = kmeans.labels_
```

```python
pune_merged = pune_merged.join(pune_df.set_index("Neighbourhoods"), on="Neighbourhoods")
```

```python
pune_merged
```

# Visualize the clusters on map of Pune

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(pune_merged['Latitude'], pune_merged['Longitude'], pune_merged['Neighbourhoods'], pune_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
